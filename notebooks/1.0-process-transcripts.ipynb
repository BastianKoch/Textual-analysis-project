# Process Earnings Call Transcripts

This notebook demonstrates how to clean and process earnings call transcripts using the `text_processing` module.

**Steps:**
1. Load and inspect a sample transcript
2. Test cleaning functions on a sample
3. Process all transcripts from `data/raw/Transcripts/` → `data/interim/transcripts_cleaned/`

## 1. Setup and Imports

Import the text processing functions from the `src/` module.

```python
from pathlib import Path
from src.text_processing import (
    load_transcript,
    extract_presentation,
    clean_text,
    process_transcript,
    process_all_transcripts
)
```

## 2. Load and Inspect a Sample Transcript

Let's load one transcript and see what it looks like.

```python
# Load a sample transcript
sample_path = Path("data/raw/Transcripts/2003/612460.txt")
raw_text = load_transcript(sample_path)

# Show first 1000 characters
print(f"Transcript length: {len(raw_text)} characters")
print(f"\nFirst 1000 characters:\n{'-'*60}")
print(raw_text[:1000])
```

## 3. Test Cleaning Functions

Extract the presentation content and clean the text. Compare before and after.

```python
# Extract only the presentation content (removes metadata)
presentation_text = extract_presentation(raw_text)

print(f"Original length: {len(raw_text)} characters")
print(f"After extraction: {len(presentation_text)} characters")
print(f"\nFirst 500 characters of presentation:\n{'-'*60}")
print(presentation_text[:500])
```

```python
# Clean the text (lowercase, remove speaker tags, etc.)
cleaned_text = clean_text(presentation_text)

print(f"After cleaning: {len(cleaned_text)} characters")
print(f"\nFirst 500 characters of cleaned text:\n{'-'*60}")
print(cleaned_text[:500])
```

## 4. Process All Transcripts

Now let's process all transcripts and save them to the interim folder. This preserves the directory structure (year folders).

```python
# Process all transcripts from raw to interim
input_dir = Path("data/raw/Transcripts")
output_dir = Path("data/interim/transcripts_cleaned")

# Run the batch processing
stats = process_all_transcripts(
    input_dir=input_dir,
    output_dir=output_dir,
    lowercase=True,
    remove_speaker_tags=True,
    remove_extra_whitespace=True
)

print(f"Processing complete!")
print(f"  Processed: {stats['processed']} files")
print(f"  Failed: {stats['failed']} files")
```

## 5. Verify Output

Check that the cleaned transcripts were saved correctly.

```python
# Load a cleaned transcript to verify
cleaned_sample = output_dir / "2003/612460.txt"
if cleaned_sample.exists():
    with open(cleaned_sample, 'r') as f:
        cleaned_content = f.read()
    
    print(f"Cleaned transcript saved successfully!")
    print(f"Length: {len(cleaned_content)} characters")
    print(f"\nFirst 300 characters:\n{'-'*60}")
    print(cleaned_content[:300])
else:
    print("Cleaned file not found!")
```

## 6. Customizing the Cleaning Process

You can customize the cleaning by passing different parameters to `clean_text()`:
- `lowercase=False` - keep original case
- `remove_speaker_tags=False` - keep speaker names
- `remove_extra_whitespace=False` - preserve formatting

Example: Process with original case preserved:

```python
# Example: Custom cleaning with case preserved
custom_cleaned = clean_text(
    presentation_text, 
    lowercase=False,  # Keep original case
    remove_speaker_tags=True,
    remove_extra_whitespace=True
)

print("Custom cleaned text (case preserved):")
print(custom_cleaned[:300])
```
